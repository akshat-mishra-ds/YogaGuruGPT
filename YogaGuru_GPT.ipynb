# **Project Summary -**

The rapid advancement of Large Language Models (LLMs) has transformed the way users access and interact with information. However, general-purpose language models often lack specialized knowledge and may provide inaccurate or incomplete responses when dealing with domain-specific topics. In the field of yoga education, learners frequently rely on multiple books and resources to understand concepts related to asanas, pranayama, yoga philosophy, anatomy, and meditation. This scattered information can make learning difficult and time-consuming.

The objective of this project is to develop **YogaGuruGPT: a domain-specific conversational assistant designed for the Education and Training industry.**

The chatbot is built using -
- Retrieval-Augmented Generation (RAG) architecture that combines the knowledge retrieval capabilities of vector databases with the natural language generation abilities of a pre-trained language model.
- Instead of training a model from scratch, the project leverages pre-trained open-source models and enhances their responses using authentic yoga literature.

The **knowledge base for YogaGuruGPT consists of six authoritative yoga texts**:
- Yoga Anatomy,
- Bhagavad Gita,
- Hatha Yoga Pradipika,
- Prana and Pranayama,
- Inside the Yoga Sutras, and
- Shiva Samhita.

These texts cover a wide range of yoga-related topics including anatomy, breathing techniques, classical yoga philosophy, meditation practices, and traditional yogic teachings. The documents are processed and converted into smaller text chunks to facilitate efficient retrieval.

To enable semantic search, **text embeddings are generated using the Sentence Transformers all-MiniLM-L6-v2 model**. These embeddings are stored in a **FAISS vector database**, which allows the system to quickly identify the most relevant document passages for a given user query. When a question is asked, relevant chunks are retrieved and supplied as context to the **TinyLlama language model**, which then generates a response grounded in the retrieved information.

A **Gradio-based web interface is implemented to provide an interactive chatbot experience**. Users can ask yoga-related questions and receive responses derived from the uploaded knowledge sources. The chatbot is capable of answering questions about pranayama, asanas, yoga philosophy, meditation, anatomy, and other educational topics within the yoga domain.

The project demonstrates how Retrieval-Augmented Generation can be used to build specialized educational assistants without requiring expensive model fine-tuning. **The developed system provides a practical solution for improving accessibility to yoga knowledge while showcasing the integration of document retrieval, vector databases, embeddings, and large language models within a real-world application.**

### Data Loading (PDFs)

The first step in building the chatbot is to load all yoga books into memory. Each PDF is processed page by page using the PyPDFLoader utility from LangChain. This converts the raw PDF files into document objects that can later be cleaned, split, and indexed for retrieval.

All selected yoga books were successfully loaded. Each page was converted into a LangChain document object. These documents will be used as the foundation for text chunking and knowledge retrieval.

### PDF pages count

```python
# Load PDF files and display page count for each PDF

documents = []

for file in os.listdir(pdf_folder):

    # Process only PDF files
    if file.endswith(".pdf"):

        # Load the PDF
        loader = PyPDFLoader(os.path.join(pdf_folder, file))
        pages = loader.load()

        # Print PDF name and number of pages
        print(f"{file}: {len(pages)} pages")

        # Add pages to the documents list
        documents.extend(pages)

# Print total pages across all PDFs
print(f"\nTotal pages loaded: {len(documents)}")
```

```python
# Display a preview of the first few loaded document pages

# Iterate through the first 5 pages (or fewer if less than 5 pages are available)
for i in range(min(5, len(documents))):

    # Print the page number
    print(f"\nPage {i+1}")

    # Print the first 200 characters of the page content
    print(repr(documents[i].page_content[:200]))
```

## ***2. Document Chunking***

Large language models cannot efficiently process entire books at once. Therefore, the documents are divided into smaller overlapping chunks.

A chunk size of 1000 characters and an overlap of 200 characters are used. The overlap helps preserve context between consecutive chunks and improves retrieval quality.


```python
# Create a text splitter to divide documents into smaller chunks

text_splitter = RecursiveCharacterTextSplitter(

    # Maximum number of characters in each chunk
    chunk_size=1000,

    # Number of overlapping characters between consecutive chunks
    chunk_overlap=200
)

# Split the loaded documents into smaller text chunks
chunks = text_splitter.split_documents(
    documents
)

# Display the total number of chunks created
print(len(chunks))
```

The documents were successfully divided into smaller chunks. These chunks will be converted into vector embeddings and stored inside the vector database.

## ***3. Create Embeddings***

Machine learning models cannot directly understand raw text. Therefore, each document chunk is converted into a numerical vector representation called an embedding.

The Sentence Transformers model `all-MiniLM-L6-v2` is used because it generates high-quality semantic embeddings while remaining computationally efficient.

```python
# Initialize the Hugging Face embedding model

embedding_model = HuggingFaceEmbeddings(

    # Pre-trained model used to convert text chunks into vector embeddings
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

Text chunks were successfully transformed into semantic vector representations. These embeddings capture the meaning of the text and enable similarity-based retrieval.

## ***4. Create FAISS Vector Database***

A FAISS vector database is created to store the generated embeddings. FAISS enables efficient similarity search, allowing the system to quickly retrieve relevant information from thousands of document chunks.

```python
# Create a FAISS vector database from the document chunks

vector_db = FAISS.from_documents(

    # Text chunks to be converted into embeddings and stored
    chunks,

    # Embedding model used to generate vector representations
    embedding_model
)
```

```python
# =================================
# Save Vector Database
# =================================

# Save the FAISS vector database to Google Drive for future use
vector_db.save_local(

    # Path where the vector database will be stored
    "/content/drive/MyDrive/Specialization projects/YogaGuruGPT/vectorstore"
)
```

## ***5. Similarity Search Testing***

```python
# Define a user query for testing document retrieval
query = "What is Pranayama?"

# Retrieve the top 3 most relevant document chunks
# based on semantic similarity to the query
results = vector_db.similarity_search(
    query,
    k=3
)

# Display the first 500 characters of each retrieved document chunk
for doc in results:
    print(doc.page_content[:1000])
```

```python
queries = [
    "What is Kapalbhati?",
    "What are the eight limbs of yoga?",
    "Explain Karma Yoga",
    "What is Matsyendrasana?",
    "What is Nadi Shodhana?"
]

for q in queries:

    print("\n")
    print("="*80)
    print("QUERY:", q)

    result = vector_db.similarity_search(
        q,
        k=1
    )

    print(result[0].page_content[:500])
```

## ***6. Create Retriever***

```python
# Create a retriever from the FAISS vector database

retriever = vector_db.as_retriever(

    # Use Maximum Marginal Relevance (MMR) to improve result diversity
    search_type="mmr",

    search_kwargs={

        # Number of document chunks to return
        "k": 7,

        # Number of candidate chunks to consider before selecting the final results
        "fetch_k": 25
    }
)
```

## ***7. LLM Integration***

TinyLlama-1.1B-Chat is used as the language model for this project. It is a lightweight, open-source conversational model that can run efficiently on Google Colab while still providing useful responses for educational applications.

```python
# Install Dependencies
# --------------------------------
!pip install -q transformers accelerate bitsandbytes
```

```python
# ===============================
# Load TinyLlama
# ===============================

# Import the Hugging Face pipeline and PyTorch libraries
from transformers import pipeline
import torch

# Initialize the TinyLlama text generation model
pipe = pipeline(

    # Specify the task as text generation
    task="text-generation",

    # Load the pre-trained TinyLlama chat model
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",

    # Use half-precision (FP16) for faster inference and lower memory usage
    torch_dtype=torch.float16,

    # Automatically select the available hardware (GPU/CPU)
    device_map="auto",

    # Maximum number of tokens to generate in the response
    max_new_tokens=256,

    # Control randomness of generated text (lower = more focused)
    temperature=0.0,

    # Enable sampling for more natural and varied responses
    do_sample=False

)
```

```python
# =================
# Test LLM
# =================

messages = [
    {
        "role": "user",
        "content": "Explain yoga in simple words."
    }
]

response = pipe(messages)

print(response[0]["generated_text"][-1]["content"])
```

## ***8. Prompt Engineering***

A custom prompt template is used to ensure that the language model generates answers only from the retrieved context. This helps reduce hallucinations and improves factual accuracy.

```python

template = """
You are YogaGuruGPT.

Answer ONLY from the supplied context.

Do not add any information that is not explicitly mentioned.

If the answer is not present in the context, say:

'I could not find this information in the YogaGuruGPT knowledge base.'

Context:
{context}

Question:
{question}

Answer in 3-5 concise sentences:
"""
```

## ***9. Manual RAG Testing***

```python
# Define a user query
query = "What are the eight limbs of yoga?"

# Retrieve the most relevant document chunks for the query
docs = retriever.invoke(query)
```

```python
# Combine the content of all retrieved document chunks into a single context

context = "\n\n".join(

    # Extract the text content from each retrieved document
    [doc.page_content for doc in docs]
)
```

```python
# Create the final prompt by inserting the retrieved context and user question

prompt = template.format(

    # Insert the retrieved document content into the prompt
    context=context,

    # Insert the user's question into the prompt
    question=query
)
```

```python
# Create a message in chat format containing the prompt

messages = [
    {
        "role": "user",
        "content": prompt
    }
]

# Generate a response using the language model
response = pipe(messages)

# Extract the generated answer from the model's output
answer = response[0]["generated_text"][-1]["content"]

# Display the final answer
print(answer)
```

## ***10. Final Chat Function***

```python
# Define a function to answer yoga-related questions

def yoga_chat(question):

    # Retrieve relevant document chunks based on the user's question
    docs = retriever.invoke(question)

    # Combine the retrieved document content into a single context
    context = "\n\n".join(
        [doc.page_content for doc in docs]
    )

    # Create a prompt using the context and the user's question
    prompt = template.format(
        context=context,
        question=question
    )

    # Format the prompt as a chat message
    messages = [
        {
            "role": "user",
            "content": prompt
        }
    ]

    # Generate a response using the language model
    response = pipe(messages)

    # Extract the generated answer from the model output
    answer = response[0]["generated_text"][-1]["content"]

    # Return the final answer
    return answer
```

## ***12.Gradio Interface***

```python
# Import the Gradio library for building the web interface
import gradio as gr
```

```python
# Create the Gradio user interface for YogaGuruGPT

interface = gr.Interface(

    # Function to process user questions and generate answers
    fn=yoga_chat,

    # Input textbox for user questions
    inputs=gr.Textbox(
        label="Ask a Yoga Question",
        lines=3,
        placeholder="Example: What is Pranayama?"
    ),

    # Output textbox to display the generated answer
    outputs=gr.Textbox(
        label="YogaGuruGPT Answer",
        lines=15
    ),

    # Application title
    title="🧘 YogaGuruGPT",

    # Brief description of the chatbot and supported topics
    description="""
    A Specialized LLM-Based Yoga Education Assistant

    Ask questions about:
    • Yoga Anatomy
    • Pranayama
    • Yoga Sutras
    • Bhagavad Gita
    • Meditation
    • Hatha Yoga
    """
)
```
